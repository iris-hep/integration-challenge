```python
import json
def merge_cert_jsons(file_paths, output_path):
    merged = {}
    for path in file_paths:
        with open(path, "r") as f:
            data = json.load(f)  # dict: run(str) -> list[[start, end], ...]
        for run, ranges in data.items():
            merged.setdefault(run, []).extend(ranges)

    # coalesce per run (optional but recommended)
    for run, ranges in merged.items():
        merged[run] = ranges

    # write as a single JSON object (use .json extension)
    with open(output_path, "w") as out:
        json.dump(merged, out, separators=(",", ":"))

    return output_path


json_16 = "./Cert_271036-284044_13TeV_Legacy2016_Collisions16_JSON.txt"
json_17 = "./Cert_294927-306462_13TeV_UL2017_Collisions17_GoldenJSON.txt"
json_18 = "./Cert_314472-325175_13TeV_Legacy2018_Collisions18_JSON.txt"
file_paths = [json_16, json_17, json_18]

output_file = "Cert_merged.txt"

merge_cert_jsons(file_paths, output_file)

print(f"Merged data written to '{output_file}'")
```
